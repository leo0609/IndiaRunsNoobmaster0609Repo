# Imports

```python
import gzip, json
import csv
```

# Preprocessing

```python
with gzip.open("candidates.jsonl.gz", "rt") as f:
	candidates = [json.loads(line) for line in f if line.strip()]
print(len(candidates))  # 100000
```

# Ranking Step

```python
# Ranking step
# TODO: Replace actual ranking logic here (currently takes first 100 as placeholder)

ranked_candidates = [
    {"candidate_id": candidates[i]["candidate_id"], "score": 100 - i}
    for i in range(100)
]
# ranked_candidates is sorted best→worst, score 100 down to 1
```

# Generate csv file

```python
output_filename = "Noobmaster0609.csv"

with open(output_filename, "w", newline="", encoding="utf-8") as csvfile:
    fieldnames = ["candidate_id", "rank", "score", "reasoning"]
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    writer.writeheader()
    for rank, candidate in enumerate(ranked_candidates, start=1):
        writer.writerow({
            "candidate_id": candidate["candidate_id"],
            "rank":         rank,
            "score":        candidate["score"],
            "reasoning":    ""   # TODO: populate once ranking logic is in place
        })

print(f"Saved {len(ranked_candidates)} candidates to '{output_filename}'")
```
