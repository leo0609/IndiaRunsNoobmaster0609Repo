# Imports

```python
import gzip, json
import csv
```

# Preprocessing

```python
with gzip.open("candidates.jsonl.gz", "rt") as f:
	candidates = [json.loads(line) for line in f if line.strip()]
print(len(candidates))  # 100000
```

# Ranking

## Job description constants and scoring functions from ranker.py

```python
import importlib, ranker
importlib.reload(ranker)  # reload so notebook picks up changes without restarting kernel
```

## Ranking Step

```python
from ranker import score_candidate

scored = []
for c in candidates:
    score, reasoning = score_candidate(c)
    if score is not None and score > 0:
        scored.append({"candidate_id": c["candidate_id"], "score": score, "reasoning": reasoning})

scored.sort(key=lambda x: x["score"], reverse=True)
ranked_candidates = scored[:100]
```

# Generate csv file

```python
output_filename = "Noobmaster0609.csv"

with open(output_filename, "w", newline="", encoding="utf-8") as csvfile:
    fieldnames = ["candidate_id", "rank", "score", "reasoning"]
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    writer.writeheader()
    for rank, candidate in enumerate(ranked_candidates, start=1):
        writer.writerow({
            "candidate_id": candidate["candidate_id"],
            "rank":         rank,
            "score":        candidate["score"],
            "reasoning":    candidate["reasoning"]
        })

print(f"Saved {len(ranked_candidates)} candidates to '{output_filename}'")
```
